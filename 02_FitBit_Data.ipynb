```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Kaggle/Fitabase Data 4.12.16-5.12.16/dailyActivity_merged.csv")
```

```python
df.head(10)
```

```python
df.Id.nunique()
```

```python
 cols = ['Id',	'ActivityDate',	'TotalSteps',	'TotalDistance',	'TrackerDistance',	'LoggedActivitiesDistance',	'VeryActiveDistance',	'ModeratelyActiveDistance',	'LightActiveDistance',	'SedentaryActiveDistance',	'VeryActiveMinutes',	'FairlyActiveMinutes',	'LightlyActiveMinutes',	'SedentaryMinutes',	'Calories']
 df = df[cols]
```

```python
df
```

```python
df.rename(columns={'ActivityDate' : 'Date'}, inplace = True)
```

```python
df
```

```python
df['TotalMinutes'] = df.VeryActiveDistance + df.FairlyActiveMinutes + df.SedentaryMinutes + df.LightlyActiveMinutes
```

```python
df
```

```python
df['TotalHours'] = round(df.TotalMinutes/60)
df
```

```python
df.info()
```

```python
df.date = pd.to_datetime(df.Date)
df
```

```python
df.info()
df
```

```python
import datetime as dt
df['DayOfWeek'] = df.date.dt.day_name()
```

```python
plt.figure(figsize = ((6,4)))
plt.hist(df.DayOfWeek, bins=7, color = 'lightskyblue', width = 0.6)
plt.grid(True)

plt.xlabel('day of week')
plt.ylabel('freq')

plt.title('freq by day of week')
plt.show()
```

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

```python
plt.figure(figsize = ((8,6)))
plt.scatter(df.TotalSteps, df.Calories, c = df.Calories)

median_steps = 	7405
median_calories = 2134

plt.axhline(median_calories, color = 'blue', label = 'Median of Calories')
plt.axvline(median_steps, color = 'red', label = 'Median of Steps')

plt.xlabel('steps')
plt.ylabel('Calories')

plt.title('Caloties by Total Steps')
plt.legend()
plt.show()
```

```python
plt.figure(figsize = ((8,6)))
plt.scatter(df.TotalHours, df.Calories, c = df.Calories)

median_hours = 	24
median_calories = 2134

plt.axhline(median_calories, color = 'blue', label = 'Median of Calories')
plt.axvline(median_hours, color = 'red', label = 'Median of Hours')

plt.xlabel('steps')
plt.ylabel('Calories')

plt.title('Caloties by Total Steps')
plt.legend()
plt.show()
```

```python
VeryActiveMinutes = df.VeryActiveMinutes.sum()
FairlyActiveMinutes = df.FairlyActiveMinutes.sum()	
LightlyActiveMinutes = df.LightlyActiveMinutes.sum()	
SedentaryMinutes = df.SedentaryMinutes.sum()

minutes = [VeryActiveMinutes,	FairlyActiveMinutes,	LightlyActiveMinutes,	SedentaryMinutes]
labels = ['VeryActiveMinutes',	'FairlyActiveMinutes',	'LightlyActiveMinutes',	'SedentaryMinutes']

plt.pie(minutes, labels = labels, autopct='%1.1f%%', explode = [0,0,0,0.2])
plt.show()
```
